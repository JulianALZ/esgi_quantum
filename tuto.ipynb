```python
import numpy as np
from qiskit import *
```

```python
# on crée un circuit quantique
circ = QuantumCircuit(3)
```

```python
# affichage du circuit
circ.draw('mpl')
```

```python
# mettons un qubit en état de superposition
# h = porte de Hadamard = superposition d'un qubit
circ.h(0)
circ.draw('mpl')
```

```python
# Intrication de 2 qubits
# cx = CNOT = Control Not = intrication de deux qubits
circ.cx(0,1)
circ.draw('mpl')
```

```python
# Intrication des qubits 1 et 2
circ.cx(1,2)
circ.draw('mpl')
```

```python
# utilisons un simultateur quantique
# on importe Aer = le module des simulateurs
# On crée notre moteur de calcul
backend = Aer.get_backend('statevector_simulator')
```

```python
# On lance le calcul quantique
job = backend.run(circ)
```

```python
# affichons le résultat
result = job.result()
```

```python
# On a simplement des vecteurs d'état
outputstate = result.get_statevector(circ, decimals=3)
print(outputstate)
```

```python
# Affichons un histogramme
from qiskit.visualization import plot_state_city
plot_state_city(outputstate)
```
